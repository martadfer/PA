```python
def associacaoQuant(ls_abordagem,covariavel):
    for cluster1 in range(len(ls_abordagem)):
        for cluster2 in range(cluster1 + 1, len(ls_abordagem)):
            resposta = researchpy.ttest(abordagem_medidas[cluster1][covariavel],abordagem_medidas[cluster2][covariavel])
            resposta[0].to_excel(DIR_ABORDAGENS / 'dCohen' / f"intervalo_conf_{covariavel}_cluster{cluster1 + 1}&cluster{cluster2 + 1}.xlsx", 
                           encoding="latin1", index=False)
            resposta[1].to_excel(DIR_ABORDAGENS / 'dCohen' / f"dCohen_{covariavel}_cluster{cluster1 + 1}&cluster{cluster2 + 1}.xlsx", 
                            encoding="latin1", index=False)
```
