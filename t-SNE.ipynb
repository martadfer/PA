```python
#bibliotecas pandas de leitura de arquivos
from sklearn.manifold import TSNE
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from numpy import reshape
import seaborn as sns
from pathlib import Path
```

```python
# Definição de diretórios e Disco de leitura dos arquivos (Desktop ou Notebook) 
config_dirs = open("prefixo_dados.txt").readlines() 
DIR_PREFIXO = Path(config_dirs[0].strip())


DIRETORIO_PRINCIPAL =  DIR_PREFIXO / "Projeto_PA" / "Projeto_PA_validado"
DIR_KMEANS = DIRETORIO_PRINCIPAL / "KMeansRandom"
DIR_TSNE = DIRETORIO_PRINCIPAL / "tSNE"
DIR_BASE = DIR_PREFIXO / "Projeto_PA" / "dataset_modificado"
```

```python
#leitura do dataset completo com dados cadastrais, medidas e estatísticas

#dataset = pd.read_csv(DIR_BASE / "pacientes_completos_tratados.csv",sep=';', encoding="Latin1")

dataset_medidas = pd.read_csv(DIR_KMEANS / f"KMEANS_approach_medidas_todos_pacientes.csv",sep=';', encoding="Latin1")
```

```python
tsne = TSNE(n_components=2, verbose=1, random_state=123)
pca = tsne.fit_transform(dataset_medidas.loc[:,'sist 9:00':'diast 8:45']) 
```

```python

df = pd.DataFrame()
df["grupo"] = dataset_medidas['grupo']
df["comp-1"] = pca[:,0]
df["comp-2"] = pca[:,1]
plt.rcParams['figure.figsize'] = [16, 8]

#plt.figure(figsize=(25,25))
sns.set(font_scale=2)
sns.set_theme(style="whitegrid")
sns.scatterplot(x="comp-1", y="comp-2", hue=df.grupo.tolist(),
                palette=['b','r','y','g','k','m','c'],
                data=df).set(title="Medidas PAS/PAD projetadas em t-SNE")
plt.savefig(DIR_TSNE / 'tSNE.png',
               bbox_inches="tight", pad_inches=1)
#plt.savefig(DIR_TSNE / "graficos" / 'tSNE' /'tSNE.png',
#               bbox_inches="tight", pad_inches=1)
    
plt.show()

```
